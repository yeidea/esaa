```python
input_fd = open('youtube_new/JPvideos.csv', errors = 'backslashreplace',encoding='utf-8')
JPvideos = pd.read_csv(input_fd)
print(JPvideos.head())
```

분석 4번. 나라별 하루에 얼마나 많이 유튜브를 보는지
한국과 일본의 한 사람당 하루 평균 유튜브 시청 횟수를 살펴본 결과, 한국이 0.78회, 일본이 0.12회로 한국이 훨씬 높은 것으로 나타났다.
일본의 인구 수가 많은 점을 고려해, 하루 평균 전국의 유튜브 스트리밍 횟수를 비교해보아도, 한국이 146.9억회, 일본이 53.8억회로, 한국이 약 3배 가량 스트리밍 수가 많음을 알 수 있다.

```python
kr100 = KRvideos.sort_values('comment_count', ascending=False).iloc[:100,:]
```

```python
kr100.head()
```

```python
kr100['tag_count'] = kr100['tags'].apply(lambda x:x.count('|')+1)
```

```python
kr100[['title','tags','views','comment_count','tag_count']]
```

댓글수 100위까지의 태그들과 태그 수를 살펴보면 어떠한 상관관계가 보이지는 않는다.

```python
kr100['ranking_group']=sorted(list(range(0,10))*10)
print(kr100.groupby('ranking_group').views.mean())
```
